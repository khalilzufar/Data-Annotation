# Data Filtering

```python
import pandas as pd

# Load the raw scraped dataset
df = pd.read_csv('GK_raw_dataset.csv')

# Drop rows where the review text is completely missing (NaN)
df = df.dropna(subset=['text'])

# Filter text containing high-frequency Indonesian anchor words and slang
kata_kunci = 'tempat|bagus|kurang|pantai|bali|indah|bgt|nyaman|bersih|rame|ramai|enak'
df_id = df[df['text'].str.contains(kata_kunci, case=False, na=False)]

# Extract the first 200 rows for the manual human annotation sample
df_sample = df_id.head(200)[['text']]

# Export to a clean baseline CSV file with explicit UTF-8 verification
df_sample.to_csv('dataset_raw_sampled.csv', index=False, encoding='utf-8')
```
